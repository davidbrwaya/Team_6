```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as py
from scipy.stats import pearsonr
from scipy.stats import linregress
import seaborn as sns

```

```python
# Fitbit Data Files
daily_activity_path = 'data/dailyActivity_merged1.csv'
sleep_day_path = 'data/sleepDay_merged.csv'
weight_log_info_path = 'data/weightLogInfo_merged.csv'
# Read the daily activity results 
daily_activity = pd.read_csv(daily_activity_path)
sleep_day = pd.read_csv(sleep_day_path)
weight_log_info = pd.read_csv(weight_log_info_path)
# Combine the data into a single DataFrame
combined_df = pd.merge(daily_activity, sleep_day, on = 'Id')
combined_df = pd.merge(combined_df, weight_log_info , on = 'Id' )
combined_df.head()
```

```python
# Build DataFrame with values related to Fitbit Fitness
user_id = combined_df['Id']
total_steps = combined_df['TotalSteps']
total_distance = combined_df['TotalDistance']
calories = combined_df['Calories']
total_minutes_asleep = combined_df['TotalMinutesAsleep']
weight_kg = combined_df['WeightKg']
weight_pounds = combined_df['WeightPounds']
total_time_in_bed = combined_df['TotalTimeInBed']
very_active_distance = combined_df['VeryActiveDistance']
very_active_minutes = combined_df['VeryActiveMinutes']
# Create a DataFrame
combined_df = pd.DataFrame({
    'Id' : user_id,
    'TotalSteps' : total_steps,
    'TotalDistance' : total_distance,
    'Calories' : calories,
    'TotalMinutesAsleep' : total_minutes_asleep,
    'WeightKg' : weight_kg,
    'WeightPounds' : weight_pounds,
    'TotalTimeInBed' : total_time_in_bed,
    'VeryActiveDistance' : very_active_distance,
    'VeryActiveMinutes' : very_active_minutes,
    
})    
# Display the first few rows of the DataFrame
combined_df.head()
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=combined_df['TotalMinutesAsleep'], y=combined_df['WeightPounds'], color='blue', alpha=0.7)
sns.regplot(x=combined_df['TotalMinutesAsleep'], y=combined_df['WeightPounds'], scatter=False, color='red')  # Add this line for regression
plt.title('Total Minutes Asleep vs. Weight (Pounds) with Regression Line')
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Weight (Pounds)')
plt.grid(True)
plt.show()
```

```python
# Calculate the sum of Total Minutes Asleep and Weight (Pounds)
total_minutes_asleep_sum = combined_df['TotalMinutesAsleep'].sum()
weight_sum = combined_df['WeightPounds'].sum()

# Plotting a pie chart
labels = ['Total Minutes Asleep', 'Weight (Pounds)']
sizes = [total_minutes_asleep_sum, weight_sum]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Proportion of Total Minutes Asleep and Weight (Pounds)')
plt.show()
```

```python
#Plot Very Active Distance Vs. Very Active Minutes
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.regplot(x=combined_df['VeryActiveDistance'], y=combined_df['VeryActiveMinutes'], color='blue', scatter_kws={'alpha':0.7})
plt.title('Very Active Distance vs. Very Active Minutes with Regression Line')
plt.xlabel('Very Active Distance (miles)')
plt.ylabel('Very Active Minutes')
plt.show()
```

```python
# Create bins for Very Active Distance
bins = [0, 2, 5, 10, float('inf')]
labels = ['0-2 miles', '2-5 miles', '5-10 miles', '10+ miles']

# Apply bins to the DataFrame
combined_df['DistanceCategory'] = pd.cut(combined_df['VeryActiveDistance'], bins=bins, labels=labels, right=False)

# Calculate the average Very Active Minutes for each category
avg_minutes_by_category = combined_df.groupby('DistanceCategory')['VeryActiveMinutes'].mean()

# Bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_minutes_by_category.index, y=avg_minutes_by_category, color='blue')
plt.title('Average Very Active Minutes by Distance Category')
plt.xlabel('Very Active Distance Category (miles)')
plt.ylabel('Average Very Active Minutes')
plt.show()
```

```python
#Plot Total Steps vs Calories
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(combined_df['TotalDistance'], combined_df['Calories'], alpha=0.5, label='Data Points')

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(combined_df['TotalDistance'], combined_df['Calories'])
line = slope * combined_df['TotalDistance'] + intercept

# Plotting the linear regression line
plt.plot(combined_df['TotalDistance'], line, color='red', label='Linear Regression Line')

plt.title('TotalDistance vs Calories with Linear Regression')
plt.xlabel('TotalDistance')
plt.ylabel('Calories')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Sorting the DataFrame by 'TotalDistance' for a meaningful line plot
sorted_df = combined_df.sort_values(by='TotalDistance')

# Line plot for Total Steps and Calories
plt.figure(figsize=(12, 6))
plt.plot(sorted_df['TotalDistance'], sorted_df['Calories'], marker='o', linestyle='-', color='blue')
plt.title('TotalDistance vs Calories')
plt.xlabel('TotalDistance')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```
